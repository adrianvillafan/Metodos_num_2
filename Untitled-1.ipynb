```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
E = np.linspace(6.98, 6.99, 2)
x1 = -4
xN = 4.5
s = 0.1
```

```python
def phi(E, valores_x, s):
  p_1 = 0
  p_2 = 1e-5
  p_3 = []
  for x in valores_x:
    p_3t = ((x*x-E)*s*s+2)*p_2-p_1
    p_1 = p_2
    p_2 = p_3t
    p_3.append(p_3t)
  return p_3
```

```python
valores_x = np.arange(x1, xN+s, s)
plt.figure(figsize=(15,10))
if type(E) == list or type(E) == np.ndarray:
  for i in E:
    fdeo = phi(i, valores_x, s)
    plt.plot(valores_x, fdeo, label=f'E={i:.2f}')
else:
  fdeo = phi(E, valores_x, s)
  plt.plot(valores_x, fdeo, color = 'yellow', label=f'E={E:.2f}')
plt.legend()
plt.suptitle("Funciones de onda para el oscilador armónico", size = 20)
plt.title(f"$x_1={x1},\ x_N={xN},\ s={s}$", size = 15)
plt.grid()
plt.show()
```
